# Raspagem com Beautiful Soap
Esse notebook é responsável por encontrar nas páginas de posts do medium, as principais features para modelagem do problema, para criar arquivos python de deploy, realizar o scrapping e prever. O **Beautiful Soap** tem seus prós e contras, a escolha se deu porque inicialmente procurava uma solução mais rápida e simples, que para outros frameworks, teria que realizar algumas *integrações*.

## Bibliotecas

```python
#Time and Regular Expression
import re
import time
from datetime import date

#BeautifulSoap and Requests
import requests as rq
import bs4 as bs4
```

```python
#Tags
queries = ['data-science', 'machine-learning', 'deep-learning', 'kaggle']
url = 'https://medium.com/tag/{query}/archive/{date}'

print(date.fromordinal(date.today().toordinal()).strftime("%Y/%m/%d"))
```

## Requests 
Raspando os posts do medium que contém as tags ditas acima, e salvando as páginas em arquivos *dados_beautifulsoap/query*

```python
now = date.today()
date_back = date.fromordinal(now.toordinal()).strftime("%Y/%m/%d")
for query in queries:
    urll = url.format(query=query, date=date_back)
    print(urll)
    response = rq.get(urll)
    
    with open("./dados_beautifulsoap/{}.html".format(query), 'w+', encoding="utf-8") as output:
        output.write(response.text)
    time.sleep(2)
```

## Obtendo as features
Vamos capturar as principais features das páginas salvas

```python
with open("./dados_beautifulsoap/data-science.html", 'r+', encoding="utf-8") as inp:
    page_html = inp.read()
    parsed = bs4.BeautifulSoup(page_html)
    
    tags = parsed.select(".streamItem")
    print(len(tags))
    
    for e in tags:
        if not e.select(".graf--title"):
            continue
        if not e.select("div > span > button"):
            palms = '0'
        else:
            palms = e.select("div > span > button")[0].text
        if not e.select("div.buttonSet.u-floatRight > a"):
            comments = '0'
        else:
            comments = e.select("div.buttonSet.u-floatRight > a")[0].text
        
        link = e.findAll("a")[3]['href']
        title = e.select(".graf--title")[0].text
        reading_times = e.select("span.readingTime")[0]['title']
        print(title, palms, comments, reading_times)
        
```
